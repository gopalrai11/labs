1. The specific energy relationship at three sections in a channel is to be studied. At the most upstream section (Section 1), the channel is trapezoidal with a base-width of 3 m and side-slopes of 3H:1V. At the downstream sections (Sections 2 and 3), the channel is trapezoidal, with a base-width of 1.8 m and side-slopes of 1.6H:1V. Sections 1 and 2 have the same bed elevation, but the bed-level at Section 3 is 0.2 m lower than the other two. Assume that head losses are negligib

```python
#Q.1
class Trapezoidal:
  def __init__(self,b, m, Q):
    self.b = b #Bed width in m
    self.m = m #Side Slope
    self.Q = Q #Discharge in m3/s
 
```

```python
#creating a object for the values
c1 = Trapezoidal(b=3, m=3, Q=5)
c2 = Trapezoidal(b=1.8, m=1.6, Q=5)
c3 = Trapezoidal(b=1.8, m=1.6, Q=5)
```

```python
#a
def Plot_energy(self):
    import numpy as np
    import matplotlib.pyplot as plt
    depths=np.arange(0,6,0.02)
    ch1 = Trapezoidal( b=self.b, m=self.m, Q=2.5)
    ch2 = Trapezoidal(b=self.b, m=self.m, Q=5)
    ch3 = Trapezoidal( b=self.b, m=self.m, Q=10)
    def Area1(self):
      A = (self.b+(self.m*depths))*(depths)
      return A
    def Specific_energy(self):
      Specific_E = depths+((self.Q**2)/(2*9.81*(Area1(self)**2)))
      return Specific_E
   
    plt.figure(figsize=(8,8))
    plt.plot(Specific_energy(ch1), depths, label='Q=2.5m3/s',color ='m')
    plt.plot(Specific_energy(ch2), depths, label='Q=5m3/s',color ='r')
    plt.plot(Specific_energy(ch3), depths, label='Q=10m3/s', color ='b')
    plt.xlim(0,10)    
    plt.xlabel('Specific Energy(m)')
    plt.ylabel('Depth(m)')
    plt.title('Specific Energy Curve')
    plt.legend()
    plt.grid()    
Plot_energy(c1)
```

```python
#c
def Plot_energy1():
    import numpy as np
    import matplotlib.pyplot as plt
    depths=np.arange(0,6,0.01)
    ch1 = Trapezoidal( b=3, m=3, Q=5)
    ch2 = Trapezoidal(b=1.8, m=1.6, Q=5)
    ch3 = Trapezoidal( b=1.8, m=1.6, Q=5)
    def Area1(self):
      A = (self.b+(self.m*depths))*(depths)
      return A
    def Specific_energy(self):
      Specific_E = depths+((self.Q**2)/(2*9.81*(Area1(self)**2)))
      return Specific_E
    
    E3 = Specific_energy(ch3) +0.2   
    plt.figure(figsize=(8,8))
    plt.plot(Specific_energy(ch1), depths, label='Section-1',color ='m')
    plt.plot(Specific_energy(ch2), depths, label='Section-2',color ='r')
    plt.plot(E3, depths, label='Section-3',color ='b')
    plt.xlim(0,10)   
    plt.xlabel('Specific Energy(m)')
    plt.ylabel('Depth(m)')
    plt.title('Specific Energy Curve for different sections for Q=')
    plt.legend()
    
    diff = lambda i : abs(i - 1.4)
    n1 = min(depths, key=diff)
    loc_1 = np.where(depths == n1)[0][0]
    E = Specific_energy(ch1)[loc_1]
    

    diff = lambda i : abs(i - E)
    n2 = min(Specific_energy(ch2), key=diff)    
    loc_2 = np.where(Specific_energy(ch2) == n2)[0][0]
    d_2 = depths[loc_2]
    print('The depth at section 2 when the depth at section 1 is 1.4: ',d_2)
    
    diff1 = lambda i : abs(i - (E+.2))
    n3 = min(E3, key=diff1)    
    loc_3 = np.where(E3 == n3)[0][0]
    d_3 = depths[loc_3]
    print('The depth at section 3 when the depth at section 1 is 1.4: ',d_3)

    plt.annotate((round(E,4),round(depths[loc_1],4)), xy=(E-1,depths[loc_1]+.1))
    plt.annotate((round(E,4),round(depths[loc_2],4)), xy=(E-1,depths[loc_2]-.1))
    plt.annotate((round(E+0.2,4),round(depths[loc_3],4)), xy=(E+.2+.1,depths[loc_3]-.1))
    plt.axvline(Specific_energy(ch1)[loc_1], linestyle = '--', color ='r')
    plt.axvline(E+.2, linestyle = '--', color ='g')
    plt.axhline(depths[loc_1], linestyle = '--', color ='b')
    plt.axhline(depths[loc_2], linestyle = '--', color ='w')
    plt.axhline(depths[loc_3], linestyle = '--', color ='g')
    plt.legend()
    
Plot_energy1()
```
