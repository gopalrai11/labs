````python
import numpy as np
import matplotlib.pyplot as plt
def integrand_a(x):
  return np.exp(-x)/x
r=30
Qw=45
T=0.9
S=0.0005
ho=110
````

````python
#a
h=[]
for i in range(1,25):
  t = i
  interval = 100000
  upper=100
  lower = (r**2)*S/(T*t)
  def trapezoidal(integrand,lower,upper,interval):
    width = (upper-lower)/interval
    k = (integrand(upper) + integrand(lower))/2
    for i in range(1,interval):
      k = k + integrand(lower + i*width)
    return width*k
  result_a = trapezoidal(integrand_a,lower,upper,interval)
  h.append(ho-Qw*result_a/(4*np.pi*T))
plt.figure(figsize=(10,6))
plt.plot(range(1,25),h)
plt.title('h(r,t) vs t')
plt.xlabel('$Time(hr)$')
plt.ylabel('$h(r,t)$')
plt.grid()
````

````python
#b
h1=[]
r1 = np.arange(1,1001)
t=24
interval = 100000
upper=100
lower = (r1**2)*S/(T*t)
def trapezoidal(integrand,lower,upper,interval):
  width = (upper-lower)/interval
  k = (integrand(upper) + integrand(lower))/2
  for i in range(1,interval):
    k = k + integrand(lower + i*width)
  return width*k
result_a = trapezoidal(integrand_a,lower,upper,interval)
h1.append(ho-Qw*result_a/(4*np.pi*T))
plt.figure(figsize=(10,6))
plt.plot(r1,h1[0])
plt.title('h(r,t) vs r')
plt.xlabel('$r(m)$')
plt.ylabel('$h(r,t)$')
plt.grid()
````
