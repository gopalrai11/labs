```python
class Trapezoidal:

  def __init__(self, b, mR, mL, Q, n, So):
    
    self.b = b 
    self.mR = mR 
    self.mL = mL 
    self.Q = Q 
    self.n = n 
    self.So = So 
     
  
  def Area(self):
    Area_T = (self.b+(self.mR*d)+(self.mL*d))*(d/2)
    return Area_T
  
  #Function to estimate Wetted Perimeter
  def Wetted_perimeter(self):
    Wetted_P = self.b + self.mR*((1+self.mR**2)**.5)*d + self.mL*((1+self.mL**2)**.5)*d 
    return Wetted_P
  
  #Function to estimate Top Width
  def Top_width(self):
    Top_w = self.b+(self.mR*d)+(self.mL*d)
    return Top_w
  
```

###Estimate area (A), wetted perimeter (P), top width (T), hydraulic radius (R), specific energy (E) and specific force (M) for a given flow depth and discharge.

```python

#taking depth equals to 1m
d=1
#Function to estimate Area
def Area(self):
  Area_T = (self.b+(self.mR*d)+(self.mL*d))*(d/2)
  return Area_T
  
  #Function to estimate Wetted Perimeter
def Wetted_perimeter(self):
  Wetted_P = self.b + self.mR*((1+self.mR**2)**.5)*d + self.mL*((1+self.mL**2)**.5)*d 
  return Wetted_P
  
  #Function to estimate Top Width
def Top_width(self):
  Top_w = self.b+(self.mR*d)+(self.mL*d)
  return Top_w
  
  #Function to estimate Hydraulic radius
def Hydraulic_radius(self):
  Hydraulic_R = self.Area()/self.Wetted_perimeter()
  return Hydraulic_R
  
  #Function to estimate Specific Energy
def Specific_energy(self):
  Specific_E = d+((self.Q**2)/(2*9.81*(self.Area()**2)))
  return Specific_E
  
  #Function to estimate Specific Force
def Specific_force(self):
  Specific_F = ((self.Q**2)/(2*9.81*(self.Area()**2)))+(self.Area()*((self.b+2*self.Top_width())/(self.b+self.Top_width()))*(d/3))
  return Specific_F
  
```

```python
def Normal_d(self):
  from scipy.optimize import fsolve
  def equation(y,n = self.n,Q = self.Q,So = self.So, b=self.b,mR = self.mR,mL = self.mL):
    return ((self.n*self.Q/(self.So**.5*((self.b+(self.mR*y)+(self.mL*y))*(y/2))))**1.5)-((self.b+(self.mR*y)+(self.mL*y))/(self.b + self.mR*((1+self.mR**2)**.5)*y + self.mL*((1+self.mL**2)**.5)*y))
  return fsolve(equation,1)  
    
    
  #Function to estimate Critcal Depth
def Critical_d(self): 
  q=self.Q/self.b
  y_critical = ((q**2)/9.81)**(1/3)
  return y_critical

print('The Normal depth of the channel is',Normal_d(channel))
print('The Critical depth of the channel is',Critical_d(channel))
```

```python

def Plot_energy(self):
    import numpy as np
    import matplotlib.pyplot as plt

    d = np.arange(0,4,.2)
    def Area1(self):
      Area_T = (self.b+(self.mR*d)+(self.mL*d))*(d/2)
      return Area_T
    def Specific_energy(self):
      Specific_E = d+((self.Q**2)/(2*9.81*(Area1(self)**2)))
      return Specific_E
    E_critical = Specific_energy(self).min()
    loc = np.where(Specific_energy(self)== E_critical)[0][0]
    d_critical = round(d[loc],3)
    
    plt.figure(figsize=(8,8))
    plt.plot(Specific_energy(self), d, label='Specific Energy', marker = 'o',color ='m')
    plt.axis(xmin=0, xmax=10)
    plt.annotate((round(E_critical,3),d_critical), xy=(E_critical,d_critical))
    plt.axvline(Specific_energy(self)[loc], linestyle = '--', color ='g')
    plt.axhline(d[loc], linestyle = '--', color ='r')
    plt.xlabel('Specific Energy')
    plt.ylabel('Depth(m)')
    plt.title('Specific Energy Curve')
    
    plt.legend()
    plt.grid()
print((Plot_energy(channel1)))    
```

```python
def Plot_profile(self):
    import matplotlib.pyplot as plt
    import numpy as np

    x=np.arange(0,500)    
    b_slope=-self.So*x
    plt.plot(x,b_slope,'black', label='Channel Bed')
    ndl=Normal_d(self)-self.So*x
    plt.plot(x,ndl,'blue', label = 'Normal Depth Line')
    cdl=Critical_d(self)-self.So*x
    plt.plot(x,cdl,'red', label = 'Critical Depth Line')
    plt.title("Longitudinal Profile of Channel")

    plt.xlabel("Length in meter")
    plt.ylabel("Depth in meter")
    plt.plot()
    plt.legend()
print(Plot_profile(channel1))    
```
