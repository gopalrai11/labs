```python
#A)#for Rhone river at lyon
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
mis_discharge = np.array(df['mississippi'].dropna()).astype(float)
#Normal
norm=stats.norm.fit(df['Rhone'])
print('Normal Distribution')
print('The fitted parameters of normal distribution for the Rhone flood discharge data:')
print('The loc is:',norm[0],'and the scale parameter is:',norm[1])
log_norm=stats.lognorm.fit(df['Rhone'])
print('Log-Normal Distribution')
print('The fitted parameters of log-normal distribution for the Rhone flood discharge data:')
print('The shape parameter is:',log_norm[0],', the log parameter is:',log_norm[1],'and the scale parameter is:',log_norm[2])
pearson=stats.pearson3.fit(df['Rhone'])
print('Pearson Type III Distribution')
print('The fitted parameters of pearson_3 distribution for the Rhone flood discharge data:')
print('The shape parameter is:',pearson[0],', the log parameter is:',pearson[1],'and the scale parameter is:',pearson[2])
gamma_dist=stats.gamma.fit(df['Rhone'])
print('Gamma Distribution')
print('The fitted parameters of Gamma distribution for the Rhone flood discharge data:')
print('The shape parameter is:',gamma_dist[0],', the log parameter is:',gamma_dist[1],'and the scale parameter is:',gamma_dist[2])
gumbel_r=stats.gumbel_r.fit(df['Rhone'])
print('Gumbel_r Distribution')
print('The fitted parameters of Gumbel_r distribution for the Rhone flood discharge data:')
print( 'The log parameter is:',gumbel_r[0],'and the scale parameter is:',gumbel_r[1])
gumbel_l=stats.gumbel_l.fit(df['Rhone'])
```

the Gumbel’s distribution parameters we computed with those in his paper are (2155.12,635.56) for the rhone river,whereas in the paper the parameters are (2177,548) so, the log parameter is under-predicting whereas the scale parameter is over-predicting.

```python
#b for rhone river
df1=pd.DataFrame()
df1['Distribution']=['Normal','Log-Normal','Pearson Type-3','Gamma','Gumbel']
df1[['Test_Statistics','Chi_Critical','AIC_score','BIC_score','Null_hypothesis at 0.05% significance']]=0
hist = plt.hist(df['Rhone'], bins= 40)
plt.close()
observed = hist[0]
bin=hist[1]
#normal distribution
cdf_fitted = stats.norm.cdf(bin,norm[0],norm[1])
calculated = []
for i in range(0,len(cdf_fitted)-1):
    if i == 0 :
      val = np.ceil(cdf_fitted[0]*len(df['Rhone']))
    else:  
      val= np.ceil((cdf_fitted[i+1]-cdf_fitted[i])*len(df['Rhone']))
    calculated.append(val)
calculated
alpha=0.05
test=[]
chi = stats.chisquare(f_obs = observed ,f_exp = calculated )
test.append(chi[0])
critical_chi = stats.chi2.ppf(q=1-alpha, df=(len(observed)-1-2))
df1['Chi_Critical'][0]=critical_chi

#log-normal distribution
cdf_fitted1 = stats.lognorm.cdf(bin,log_norm[0],log_norm[1],log_norm[2])
calculated1 = []
for j in range(0,len(cdf_fitted1)-1):
    if j == 0 :
      val1 = np.ceil(cdf_fitted1[0]*len(df['Rhone']))
    else:  
      val1= np.ceil((cdf_fitted1[j+1]-cdf_fitted1[j])*len(df['Rhone']))
    calculated1.append(val1)

chi1 = stats.chisquare(f_obs = observed ,f_exp = calculated1 )
test.append(chi1[0])

#Pearson type-3 distribution
cdf_fitted2 = stats.pearson3.cdf(bin,pearson[0],pearson[1],pearson[2])
calculated2=[]
for k in range(0,len(cdf_fitted1)-1):
    if k == 0 :
      val2= np.ceil(cdf_fitted2[0]*len(df['Rhone']))
    else:  
      val2= np.ceil((cdf_fitted2[k+1]-cdf_fitted2[k])*len(df['Rhone']))
    calculated2.append(val2)

chi2 = stats.chisquare(f_obs = observed ,f_exp = calculated2 )
test.append(chi2[0])
critical_chi1 = stats.chi2.ppf(q=1-alpha, df=(len(observed)-1-3))
df1['Chi_Critical'][1:4]=critical_chi1
#Gamma distribution
cdf_fitted3 = stats.gamma.cdf(bin,gamma_dist[0],gamma_dist[1],gamma_dist[2])
calculated3=[]
for l in range(0,len(cdf_fitted3)-1):
    if l == 0 :
      val3= np.ceil(cdf_fitted3[0]*len(df['Rhone']))
    else:  
      val3= np.ceil((cdf_fitted3[l+1]-cdf_fitted3[l])*len(df['Rhone']))
    calculated3.append(val3)

chi3 = stats.chisquare(f_obs = observed ,f_exp = calculated3 )
test.append(chi3[0])
#gumbel distribution
cdf_fitted4= stats.gumbel_r.cdf(bin,gumbel_r[0],gumbel_r[1])
calculated4=[]
for m in range(0,len(cdf_fitted4)-1):
    if m == 0 :
      val4= np.ceil(cdf_fitted4[0]*len(df['Rhone']))
    else:  
      val4= np.ceil((cdf_fitted4[m+1]-cdf_fitted4[m])*len(df['Rhone']))
    calculated4.append(val4)

chi4 = stats.chisquare(f_obs = observed ,f_exp = calculated4 )
test.append(chi4[0])
critical_chi = stats.chi2.ppf(q=1-alpha, df=(len(observed)-1-2))
df1['Chi_Critical'][4]=critical_chi
df1['Test_Statistics']=test
df1['Null_hypothesis at 0.05% significance'] = df1['Test_Statistics'].apply(lambda x: 'Can Not be rejected' if x <= df1['Chi_Critical'][0] else 'Can be rejected')

```

```python
def aic(dist):
  x = getattr(stats,dist)
  y = x.fit(df['Rhone'])
  if len(y) == 3:
    shape,location,scale = y
    log_l = x.logpdf(df['Rhone'],shape,location,scale).sum() 
    AIC = 2*len(y)- 2*(log_l)
    BIC =len(df['Rhone'])*len(y) - 2*(log_l)
  else:
    location,scale = y
    log_l = x.logpdf(df['Rhone'],location,scale).sum()
    AIC = 2*len(y)- 2*(log_l)
    BIC =len(df['Rhone'])*len(y) - 2*(log_l)
  return AIC,BIC
dist=['norm','lognorm','pearson3','gamma','gumbel_r']
Aic=[]
Bic=[]
for i in dist:
  a=aic(i)[0]
  b=aic(i)[1]
  Aic.append(a)
  Bic.append(b)
df1['AIC_score']=Aic
df1['BIC_score']=Bic
c=Aic.index(min(Aic))
d=Bic.index(min(Bic))
print('FOR RHONE RIVER')
print('According to AIC creteria ',dist[c],' distribution best fits the data.')
print('According to BIC creteria ',dist[d],' distribution best fits the data.')
df1
```
